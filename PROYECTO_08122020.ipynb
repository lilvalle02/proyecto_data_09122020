Introduction:
Start up creada en California en 2012 cuyo modelo de negocios consiste en un “peer to peer lending” /prestamos entre particulares, sus productos están catalogados ante la SEC como “valores” , Lending Club permite a sus prestatarios crear listas de prestamos en su plataforma aportando su información personal y el financiamiento que buscan , se tratan de préstamos entre particulares que rondan los montos de $10,000 USD a $ 45,000 USD.
Lending Club es el encargado de asignar un score a los usuarios y los financiamientos que necesitan, y en caso de contar con un buen score, Lending Club ® es el encargado de determinar las tasas de interés, el tipo de interés.   

*Las bases descargadas de Kaggel contienen la información de todos los créditos otorgador por Lending Club del periodo de 2015 al 2017, las hipótesis se dividen en el desde la perspectiva del cliente y desde la perspectiva de la cartera /portafolio manejado por Lending Club®
